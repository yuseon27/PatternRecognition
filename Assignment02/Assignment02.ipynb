```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sympy import *
from scipy.misc import derivative
```

# Define a differential function :   $ f(x) = \frac{x}{e^{x}+1} $

```python
def func(x) :
    function = x / (np.exp(x) + 1)
    return function
```

# Define domain of the function: $ x = (-2, 2) $

```python
x = np.arange(-2, 2, 0.1)
```

# Plot

```python
plt.plot(x, func(x))
plt.show()
```

# Select a point within the domain

```python
point = random.choice(x)
point
```

# Mark the selected point on the function

```python
plt.plot(x, func(x))
plt.plot(point, func(point), marker='o', ms=5)
plt.show()
```

# Define first-order Taylor approximation at the selected point

```python
def taylor_approximation(domain, point) :
    taylor = func(point) + derivative(func, point, dx=1e-5)*(domain - point)
    return taylor
```

# Plot Taylor approximation with the same domain of the original function

```python
plt.plot(x, func(x))
plt.plot(point, func(point), marker='o', ms=5)
plt.plot(x, taylor_approximation(x, point))
plt.show()
```
